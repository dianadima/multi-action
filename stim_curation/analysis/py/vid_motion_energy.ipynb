```python
import moten
import numpy as np
import glob
from scipy.io import savemat
```

```python
video_sets = [sorted(glob.glob('/Users/dianadima/Desktop/MultiAction/stimuli/*.mp4'))]
nframes = [40]
savenames = ['full_videoset']
```

```python
def motion_energy(videoname, nframes, savename):

    # Stream and convert the RGB video into a sequence of luminance images
    luminance_images = moten.io.video2luminance(videoname, nimages=nframes)

    # Create a pyramid of spatio-temporal gabor filters
    nimages, vdim, hdim = luminance_images.shape
    pyramid = moten.get_default_pyramid(vhsize=(vdim, hdim), fps=24)

    # Compute motion energy features
    moten_features = pyramid.project_stimulus(luminance_images)
    return moten_features
```

```python
for i in range(1):
    
    print('Running set', i+1)
    vset = video_sets[i]
    nf = nframes[i]
    savename = savenames[i]
    feat = []
    
    for v in vset: 
        
        feat.append(motion_energy(v,nf,savename))
    
    #Save output to .mat
    outpath = f"/Users/dianadima/Desktop/MultiAction/{savename}_motion_energy.mat"
    savemat(outpath, {'feat':feat})
        
```

```python
import matplotlib.pyplot as plt

print(feat[0].shape)
fig, ax = plt.subplots(figsize=(12, 12))
ax.matshow(feat[0], aspect='auto')
plt.show()
```
